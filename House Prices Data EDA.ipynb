```python
df_highCorr = data.corr()
df_highCorr = df_highCorr[abs(df_highCorr['SalePrice']) > .25]
df_highCorr = df_highCorr[[i for i in df_highCorr.index]]
print(len(df_highCorr))
# so filter the data for the heatmap to graph only the cols with some corr, don't need the ones
# with 0 or very low correlation (white ones)
```

```python
plt.figure(figsize=(16, 10))
corr = df_highCorr
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(10, 290, s=80, l=55, n=100),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
